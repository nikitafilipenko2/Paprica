```python
import trimesh
import matplotlib.pyplot as plt

mesh = trimesh.load_mesh("gbt-p7a-700.stl")

print(mesh)
print(f"Vertices: {len(mesh.vertices)}")
print(f"Faces: {len(mesh.faces)}")
print(f"Watertight: {mesh.is_watertight}")
print(f"Manifold: {mesh.is_winding_consistent}")

mesh.show()
```

```python
vertices = mesh.vertices  # numpy-массив формы (V, 3)
print(f"Количество уникальных вершин: {len(vertices)}")
print("Пример координат:")
print(vertices[:5])
```
